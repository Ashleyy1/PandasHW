```python
#percentage and count of male players 
len(df["Gender"])
m_df = df[df['Gender'] == 'Male']
percentage_m_df = len(m_df) / len(df["Gender"]) *100
print(percentage_m_df)
print(len(m_df))
```

```python
#percentage and count of female players
f_df = df[df['Gender'] == 'Female']
percentage_f_df = len(f_df) / len(df["Gender"]) *100
print(percentage_f_df)
print(len(f_df))
```

```python
#perfentage and count of other/ non- disclosed 
other_df = df[df['Gender'] == "Other / Non-Disclosed"]
percentage_other_df = len(other_df) / len(df["Gender"]) *100
print(percentage_other_df)
print(len(other_df))
```

```python
#DateFrame for gender demographics  
gender_dem = df.groupby('Gender').agg({
    "Percentage" : [83.58]
})
gender_dem
```

```python
#purchase analysis by gender 
purchase_analysis = df.groupby('Gender').agg({
    "Price" : ['count', 'sum', 'mean'],
    "SN" : 'nunique'
})
purchase_analysis

purchase_analysis.columns = [' '.join(col).strip()for col in purchase_analysis.columns.values]
# purchase_analysis['Average Purchase Total per Person by Gender'] = purchase_analysis['Price'/ 'Sum']
purchase_analysis
```
